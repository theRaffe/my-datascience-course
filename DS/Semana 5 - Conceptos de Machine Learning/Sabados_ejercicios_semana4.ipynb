```python
diabetes_corr_matrix = diabetes_df.corr().round(2)
sns.heatmap(data=diabetes_corr_matrix, annot=True)
sns.pairplot()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
```

```python
lm = LinearRegression()
x= diabetes_df.drop('y_target', axis = 1)
y= diabetes_df['y_target']
y = diabetes_df.loc[:, ['y_target']].values
y = StandardScaler().fit_transform(y)

train_x, test_x, train_y, test_y = train_test_split(x, y, test_size=0.3,
                                                    random_state=1)
```

```python
lr=lm.fit(train_x, train_y)
```

```python
#convert x into dataframe
plt.figure(figsize= (15,10))
ft_lm = pd.Series(lr.coef_ ,index = x.columns)
ft_lm.plot(kind='barh')
plt.show();
```

```python
print("R2 Regresion linear para Diabetes es -")
np.round(lr.score(test_x,test_y)*100,2)
```

```python
#la regresion lineal no funciona para esta caso, no hay correlacion clara
predictions = lm.predict(test_x)
plt.scatter(test_y,predictions)
```

```python
predict_test_lm = lr.predict(test_x)
df = pd.DataFrame({'Actual': test_y, 'Predicted': predict_test_lm}) 
df
```

```python
import numpy as np
from sklearn import metrics
```

```python
print("MSE de test para LR -")
np.round(metrics.mean_squared_error(test_y,predict_test_lm),2)
```
