```python
#Hagan un codigo que reciba una lista de palabras y calcule la frecuencia de los caracteres
from functools import reduce
m_list = ["Hola", "Mundo", "como", "estas"]

def m_reduce(x,y):
    return list(x) + list(y)
key_dict = set(reduce(m_reduce, m_list))
str_list = ''.join(m_list)


{a_key : str_list.count(a_key)   for a_key in key_dict}
```

```python
#Beautiful soap
import requests
from bs4 import BeautifulSoup

URL_Mercado_Libre = 'https://celulares.mercadolibre.com.mx/samsung/#origin=brand_carousel&position=1&id=206'
response_mercado = BeautifulSoup(requests.get(URL_Mercado_Libre).text, "lxml")
```

```python
import pandas as pd
```

```python
import re


all_product_item = response_mercado.find_all('li', class_="results-item")

def get_tag_value(object_tag, tag, class_name):
    #print(f'object_tag {object_tag}')
    item = object_tag.find(tag, class_=class_name)
    if item and item.get_text() != '':
        return item.get_text().strip()
    return None
    
def get_product_object(object_tag):
    title = get_tag_value(object_tag,'span', 'main-title')
    price = get_tag_value(object_tag,'span', 'price__fraction')
    price_old = get_tag_value(object_tag,'span', 'price-old')
    sells_item = get_tag_value(object_tag,'div', 'item__condition')
    sells_arr = sells_item.split(' ')
    sells_ = sells_arr[0]  if len(sells_arr) > 1 else '0'
    
    price_old = re.sub("[^\d\.]", "", str(price_old))
    price_old = price_old if price_old != '' else '0'
    price = re.sub("[^\d\.]", "", str(price))
    price_item = price_old if price_old != '' else price
    price_desc = price if price_old != '0' else price_old
    
    return {
        "title" : title,
        "price" : float(price_item),
        "sells" : int(sells_),
        "price_desc" : price_desc
    }
    
#for a_product_item in all_product_item:
#    print(a_product_item.find('span', class_="main-title").get_text().strip())

list_product = [get_product_object(a_prod_item) for a_prod_item in all_product_item]
df_product = pd.DataFrame(list_product)
df_product
```

```python
df_product['price_usd'] = df_product['price']  * 20
df_product
```
