## 2. Archivo *ratings*

### 1. ¿Conocemos los datos? 

```python
#Lectura de los datos 
data2=pd.read_csv('movielens/ratings.csv')

# Creacion de los dataframes 
df2=pd.DataFrame(data2)

# Encabezados 
df2.head()

# Existen datos del usuario que las califico, las peliculas calificadas por el, la calificación y la hora en que fueron calificadas
```

### 2. Obtengamos los estadísticos básicos de las calificaciones 


```python
#Principales estadísticos descriptivos
df2['rating'].describe()

# Count - Cuantos registros existen
# Mean  - Promedio
# Std   - Desviación estándar 
# min   - Valor Mínimo
# 25%   - 25 percentil 
# 50%   - 50 percentil 
# 75%   - 75 percentil 
# max   - Valor máximo

```

### 3. Cambiemos la escala de las calificaciones de un máximo de 5 a 100


```python
# Usando la regla de 3 tenemos que la nueva columna (NC) respecto a la vieja (VC) se calcula como
#       NC= 100*VC/5

df2['NC']=100*df2['rating']/5

df2.head()
```

### 4. ¿Cuántas calificaciones tiene cada película? ¿Cuál es la calificación media de cada uno?

```python
# Agrupamos las columnas de rating y movieId en terminos del movie Id para obtener 
Grouped=df2[['movieId','rating']].groupby('movieId')
```

```python
Grouped.count()[:10]
```

```python
Grouped.mean()[:10]
```

### 5. ¿cuáles son las películas son las 20 mejores  y peores según las calificaciones? ¿Cuáles son las más calificadas? 

```python
# Las 20 mejores peliculas 
# Contamos los datos agrupados . Los ordenamos siguiendo el rating  de forma descendete. traemos las pimeras 10 columnas
Grouped.mean().sort_values(by='rating',ascending=False).iloc[:10]
```

```python
# Las 20 peores peliculas 
# Contamos los datos agrupados . Los ordenamos siguiendo el rating  de forma descendete. traemos las pimeras 10 columnas
Grouped.mean().sort_values(by='rating',ascending=False).iloc[-10:]
```

```python
# Peliculas más calficadas
# Contamos los datos agrupados . Los ordenamos siguiendo el rating  de forma descendete. traemos las pimeras 10 columnas
Grouped.mean().sort_values(by='rating',ascending=False).iloc[:10]
```

## 3. Pongamonos serios

### 1. La columna de genero en el DataFrame tiene divisiones, ¿podemos poner cada genero en una columna? 

```python
g=df1['Genero'].str.split('|',n=5,expand=True)
g.columns=['G1','G2','G3','G4','G5','G6']
g
```

### 2. Concatena las nuevas columnas y quita la columna vieja

```python
df1=pd.concat([df1, g], axis=1)
df1.drop(columns=['Genero']).head()
```

### 3. Agrega al dataframe pasado una columna de las calificaciones promedio

```python
# Cambiemos los nombres de las columnas del segundo dataframe
df2.columns=['UID','ID','Cal','time','NC']

# Creamos la columna de la media
rat=df2[['ID','Cal']].groupby('ID').mean()
```

```python
# Colocamos las mediciones a los IDs repectivos 
df1=df1.merge(rat,on='ID',how='inner')
        
```
