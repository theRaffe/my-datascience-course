```python
import requests
import shutil
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np

m_url = 'https://animeclubgt.com/wp-content/uploads/2016/08/198.jpg'
m_response = requests.get(m_url, stream = True)
with open('perro.jpg', 'wb') as out_file:
    shutil.copyfileobj(m_response.raw, out_file)
del m_response
```

```python
img = mpimg.imread('perro.jpg')
plt.imshow(img)
plt.show()
```
