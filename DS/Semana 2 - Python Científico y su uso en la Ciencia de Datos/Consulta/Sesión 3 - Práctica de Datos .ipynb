1. Descargar las primeras 4 páginas
2. Agregar columna de género
3. Agregar columna de entidad

````python
import json
import time

#m_df = pd.DataFrame(declarantes_json.json())
political_parties = ['PRI']
#['PRI', 'PAN', 'PT', 'PVEM', 'PRD', 'MC', 'Nueva Alianza', 'Morena', 'PH', 'PES']
all_genres = ['M','F']
all_pages = [1,2,3,4]
all_requests = np.array(np.meshgrid(all_genres,all_pages,political_parties)).T.reshape(8,3)
print(all_requests)
df_genre_party = pd.DataFrame()
ls_genre_party = []
df_total = pd.DataFrame()
for a_request in all_requests:
    a_genre = a_request[0]
    a_page = int(a_request[1])
    a_political_party = a_request[2]
    endpoint_req = f'https://3de3.mx/declarantes.json?page={a_page}&by_gender={a_genre}&by_political_party={a_political_party}'
    m_secs = np.random.randint(2,8, size=1)[0]
    time.sleep(m_secs)
    print(f'endpoint:{endpoint_req}')
    declarantes_json = requests.get(endpoint_req).json()
    
    if len(declarantes_json) > 0:
        df = pd.DataFrame(declarantes_json)
        df['genre'] = a_genre
        df['political_party'] = a_political_party
        ls_genre_party.append(df)
    
df_total = pd.concat(ls_genre_party)
df_total

````
